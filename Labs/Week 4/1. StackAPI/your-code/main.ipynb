# StackAPI

#### Import the necessary libraries here:

```python
import requests
from stackapi import StackAPI
from datetime import datetime
import pandas as pd
pd.set_option('display.max_columns', None)
```

#### Question 1: Find the questions and answers of last month.

```python
fromdate = int(datetime.timestamp(datetime(2020, 2, 1)))
todate = int(datetime.timestamp(datetime(2020, 2, 29)))
```

```python
stack = StackAPI('stackoverflow')
stack
questions = stack.fetch('questions', fromdate=fromdate, todate=todate)
```

```python
pd.json_normalize(questions['items'])
```

```python
answers = stack.fetch('answers', fromdate=fromdate, todate=todate)
pd.json_normalize(answers['items'])
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
fromdate = int(datetime.timestamp(datetime(2020, 3, 18)))
todate = int(datetime.timestamp(datetime(2020, 3, 19)))
```

```python
questions = stack.fetch('questions', fromdate=fromdate, todate=todate, tagged='python', min=5, sort='votes')
```

```python
question_python = pd.json_normalize(questions['items'])
question_python.max()
```
