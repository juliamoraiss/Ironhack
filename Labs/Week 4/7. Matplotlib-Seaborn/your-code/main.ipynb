#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y = x**2
z = math.e**x

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,8))

ax1.plot(x, y, color='red')
ax2.plot(x, z, lw=1.2)

ax1.set_xscale('linear')
ax2.set_yscale('log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
y = x**2
z = math.e**x

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,4))

ax1.plot(x, y)
ax2.plot(x, z)

ax1.set_xscale('linear')
ax2.set_yscale('log')

ax1.legend(['linear'])
ax2.legend(['log'])
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
med_steps = fitbit.groupby(by='Months_encoded')['Steps'].agg('median')
```

```python
med_week = fitbit.groupby(by='Work_or_Weekend')['Steps'].agg('median')
```

```python
# MEDIAN STEPS BY MONTH_ENCODED

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,4))

plt.sca(axes[0])
med_steps.plot(kind='line', grid=True)
plt.ylabel('Median number of steps')
plt.xlabel('Months')
plt.title('Median Steps walked by month')
plt.xticks(np.arange(13),['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])




# MEDIAN STEPS BY WORK_OR_WEEKEND

plt.sca(axes[1])
med_week.plot(kind='bar', alpha=0.8, grid=True, colormap='Paired')
plt.title('Median Steps walked by Weekend and Workdays')
plt.ylabel('Median number of steps')
plt.xlabel('Workdays or Weekend')
plt.xticks(np.arange(2),['Weekend', 'Workdays'])


plt.savefig('figures/steps_months_week.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
col = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,8))

for i in range(3):
    plt.sca(axes[i])
    plt.scatter(fitbit.loc[:, col[i]], fitbit['Steps'], alpha=0.9, edgecolors='black')
    plt.ylabel('Steps')
    plt.xlabel(col[i])
```

#### How does the bin size affect your plot?

If you have too many bins, then the data distribution will look rough, and it will be difficult to discern the signal from the noise. On the other hand, with too few bins, the histogram will lack the details needed to discern any useful pattern from the data.
